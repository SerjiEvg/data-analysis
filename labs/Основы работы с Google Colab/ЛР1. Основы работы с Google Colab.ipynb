–ü—Ä–∏—à–ª–æ –≤—Ä–µ–º—è —Å–æ–∑–¥–∞—Ç—å –µ–µ —Å –ø–æ–º–æ—â—å—é —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–π –∫–æ–º–∞–Ω–¥—ã SQL ‚Äî CREATE TABLE. –ï—Å–ª–∏ —Ç–∞–±–ª–∏—Ü–∞ —É–∂–µ —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–ª–∞ –≤ –±–∞–∑–µ –¥–∞–Ω–Ω—ã—Ö, –ø–æ—è–≤–∏—Ç—Å—è —Å–æ–æ–±—â–µ–Ω–∏–µ –æ–± –æ—à–∏–±–∫–µ. –ö—Ä–æ–º–µ —Ç–æ–≥–æ, –º—ã —É—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º ***PRIMARY KEY*** –¥–ª—è USERID, —á—Ç–æ–±—ã –ø—Ä–µ–¥–æ—Ç–≤—Ä–∞—Ç–∏—Ç—å –≤—Å—Ç–∞–≤–∫—É –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –º–æ–¥—É–ª–µ–π –∑–∞–ø–∏—Å–∏ –≤ —Ç–∞–±–ª–∏—Ü—É.
