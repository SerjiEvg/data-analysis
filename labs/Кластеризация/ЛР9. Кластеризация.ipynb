# Лабораторная работа №9 Кластеризация

`Кластеризация в машинном обучении` - это задача выбора объектов для групп (кластеров), в результате чего один объект внутри кластера был максимально похож на другой, объекты из разных кластеров были как можно более различны. Кластеризация часто используется для анализа данных и наблюдений за большими объемами информации.

## 2. Алгоритмы кластеризации

Есть много алгоритмов кластеризации на выбор:


### `entroid-based (на основе центроида)`

Эти типы алгоритмов разделяют точки данных на основе нескольких центроидов. Каждая точка данных назначается кластеру на основе ее квадрата расстояния от центроида.

Это наиболее часто используемый тип кластеризации. `Алгоритм K-средних` является одним из алгоритмов кластеризации на основе центроидов. Здесь k — количество кластеров и гиперпараметр алгоритма.


### `Hierarchical-based (Connectivity-based) (иерархический)`

Идея основана на основной идее о том, что объекты больше связаны с близлежащими объектами, чем с объектами, находящимися дальше.

Алгоритм строит дерево кластеров, поэтому все организовано сверху вниз. Первоначально каждая точка данных рассматривается как отдельный кластер. На каждой итерации аналогичные кластеры сливаются с другими кластерами, пока не образуется один кластер или K кластеров.


### `Density-based (на основе плотности)`

Данные сгруппированы по областям с высокой концентрацией точек данных, окруженных областями с низкой концентрацией точек данных. По сути, алгоритм находит места с большим количеством точек данных и вызывает эти кластеры.

Алгоритмы кластеризации этого типа не пытаются назначать выбросы кластерам, поэтому они игнорируются.

### `Distribution-based (на основе распределений)` 

Это модель кластеризации, в которой данные подгоняются по вероятности того, как они принадлежат одному и тому же распределению.

Установливается центральная точка. По мере увеличения расстояния точки данных от центра вероятность того, что она является частью этого кластера, уменьшается.

Эта модель хорошо работает с синтетическими данными и кластерами разного размера.

Метод страдает от переобучения, если не наложены ограничения на сложность модели. Более сложная модель, как правило, лучше объясняет данные, что затрудняет выбор подходящей сложности модели.
