# Начало работы с библиотекой Keras

**Глубокое обучение** — это область искусственного интеллекта, связанная с решением очень сложных задач путем имитации работы человеческого мозга. 

В глубоком обучении мы используем нейронные сети, которые используют несколько операторов, размещенных в узлах, чтобы помочь разбить проблему на более мелкие части, каждая из которых решается индивидуально. Нейронные сети могут быть очень сложными в реализации и об этой проблеме заботится **Keras - фреймворк глубокого обучения**.

## 1. Введение в Keras

**Keras** — это высокоуровневый API глубокого обучения, разработанный Google для реализации нейронных сетей. Он написан на Python и используется для упрощения реализации нейронных сетей. Он также поддерживает несколько внутренних вычислений нейронной сети.                         

Keras относительно прост в освоении и работе, потому что он предоставляет **интерфейс Python с высоким уровнем абстракции**, имея при этом возможность использования нескольких серверных частей для вычислительных целей. Это делает Keras медленнее, чем другие фреймворки глубокого обучения, но чрезвычайно удобным для начинающих. 

Высокоуровневое Keras API управляет тем, как мы создаем модели, определяем слои или настраиваем модели ввода-вывода. На этом уровне Keras также компилирует нашу модель с функциями потерь и оптимизатора, запускает процесс обучения с функцией подгонки. 

**Keras не обрабатывает низкоуровневый API**, такой как создание вычислительного графа, создание тензоров или других переменных, потому что он обрабатывается «**бэкэнд-движком**». Таким образом, Keras не поддерживает низкоуровневые вычисления. Вместо этого для этого используется другая библиотека. Таким образом, Keras — этооболочка, способная работать поверх **TensorFlow**, **CNTK** или **Theano**.

Официальную документацию по Keras можно найти [здесь](https://keras.io/).

## 6. Модели в Keras

Основная структура в Keras — это **модель**, которая определяет полный **граф сети**. А самый простой тип модели — **последовательная модель (Sequential model)**. Это линейный стек слоев.

Кроме Sequential model, в библиотеке Keras есть и другие **типы моделей**, например:

- **Functional API**: позволяет создавать более сложные модели с нелинейными связями между слоями, а также модели с несколькими входами и выходами.

- **Model subclassing**: позволяет создавать модели с полностью настраиваемой логикой обучения, что делает этот тип модели более гибким, но требует больше кода.

Каждый тип модели имеет свои преимущества и недостатки, и выбор конкретного типа зависит от задачи, которую вы пытаетесь решить.

Для более сложных архитектур следует использовать **функциональный API Keras**, позволяющий строить произвольные графы слоев.

**Input Layer (входной слой)** в Keras - это первый слой в нейронной сети, который принимает на вход данные и передает их на следующий слой для дальнейшей обработки.

В Input Layer задается форма входных данных, то есть размерность тензора, который будет поступать на вход модели. Например, если вы работаете с изображениями размером 28x28 пикселей, то форма входных данных будет (28, 28, 1), где 1 - это количество каналов (в данном случае черно-белые изображения).

Пример создания Input Layer в Keras:

```python
from keras.layers import Input

input_layer = Input(shape=(28, 28, 1))
```

Здесь мы создаем Input Layer с формой входных данных (28, 28, 1), то есть для обработки черно-белых изображений размером 28x28 пикселей.

**Hidden Layer (скрытый слой)** в Keras - это слой нейронной сети, который находится между входным и выходным слоями и выполняет обработку данных. Количество скрытых слоев и количество нейронов в каждом слое являются **гиперпараметрами**, которые могут быть настроены для достижения лучшей производительности модели.

Каждый нейрон в скрытом слое принимает на вход взвешенную сумму значений, которые поступают на вход слоя, и применяет к этой сумме нелинейную функцию активации. Нелинейность функции активации позволяет модели изучать более сложные зависимости между входными и выходными данными.

Пример создания скрытого слоя с функцией активации ReLU в Keras:

```python
from keras.layers import Dense

hidden_layer = Dense(units=64, activation='relu')(input_layer)
```

Здесь мы создаем скрытый слой с 64 нейронами и функцией активации **ReLU**, который принимает на вход **input_layer**.

**Output Layer** в Keras - это слой, который определяет выходные данные модели. Он определяет формат выходных данных и используется для решения задач классификации, регрессии или других задач машинного обучения. 

В зависимости от задачи, Output Layer может содержать **различное количество нейронов** и использовать различные функции активации. 

Например, для задачи бинарной классификации можно использовать слой с одним нейроном и функцией активации **sigmoid**, а для задачи многоклассовой классификации - слой с количеством нейронов, соответствующим числу классов, и функцией активации **softmax**.

Посмотрим как можно построить простую нейронную сеть в Keras. Мы построим сеть, которая принимает на вход **два числа**, пропускает их через скрытый слой из **10 нейронов** и выводит **одно число без ограничений**.

Выход без ограничений можно получить, избегая установки функции активации в выходном слое. Это полезно для таких задач, как **регрессия**, когда мы хотим, чтобы наш вывод мог принимать любое значение.

image.png

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

# Create a Sequential model
model = Sequential()

# Add an input layer and a hidden layer with 10 neurons
model.add(Dense(10, input_shape=(2, )))

# Add a 1-neuron output layer
model.add(Dense(1))

# Summarize your model
model.summary()
```

```python
keras.utils.plot_model(model, "my_first_model.png")
```

# 3. Cлои Keras

Keras состоит из различных типов слоев, которые являются основой строительных блоков Keras. В этом разделе мы обсудим несколько часто используемых слоев в Keras.

**Последовательная модель (Sequential model)** представляет собой линейный набор слоев. Мы можем создать последовательную модель, передав список экземпляров слоя конструктору следующим образом:

```python
from keras.models import Sequential

from keras.layers import Dense, Activation, Conv2D, MaxPooling2D, Flatten, Dropout

model = Sequential()
```

Мы также можем просто добавить слои с помощью **метода .add()** следующим образом:

```python
model = Sequential()

model.add(Dense(32, input_dim=784))

model.add(Activation('relu'))
```


Более подробное обсуждение **модели Sequential** можно найти [здесь](https://keras.io/getting-started/sequential-model-guide/)

Используя класс Sequential, можно складывать слои различных типов один за другим для создания нейронной сети. В **Keras Sequential API** доступно множество типов слоев.

### 3.1 Convolutional Layer (сверточный слой)

**Convolutional Layer** в Keras - это слой, используемый в **сверточных нейронных сетях** для обнаружения локальных пространственных шаблонов в изображениях. 

Он принимает на вход **тензор**, представляющий изображение, и применяет к нему фильтры, которые сканируют изображение, выделяя признаки и создавая карты признаков. 

Каждый фильтр представляет собой **матрицу весов**, которая перемещается по изображению с определенным шагом (шаг сдвига), выполняя **операцию свертки**. 

Результатом работы Convolutional Layer является новый тензор, содержащий карты признаков, которые могут быть использованы для классификации изображений или выполнения других задач компьютерного зрения.

![image](https://dennybritz.com/img/Convolution_schematic.gif)

Пример сверточного слоя в качестве входного слоя с входной формой 320x320x3, с 48 фильтрами размером 3x3 и использованием ReLU в качестве функции активации.

```python
input_shape=(320, 320, 3) #это входная форма изображения 320x320x3

model.add(Conv2D(48, (3, 3), activation='relu', input_shape= input_shape))
```

Другой пример выглядит следующим образом:

```python
model.add(Conv2D(48, (3, 3), activation='relu'))
```

### 3.2. Слой MaxPooling

**MAXPooling Layer** в Keras - это слой, используемый в сверточных нейронных сетях для уменьшения размерности карт признаков, полученных после применения Convolutional Layer. 

Он работает путем разбиения карты признаков на непересекающиеся прямоугольные области (обычно 2x2), и выбора максимального значения в каждой области. Это позволяет снизить количество параметров в модели, сократить время обучения и уменьшить риск переобучения. 

Результатом работы MAXPooling Layer является новая карта признаков, которая содержит только наиболее значимые признаки из исходной карты.

image.png

Чтобы уменьшить входное представление, можно использовать **MaxPool2d** и указать **размер ядра**.

```python
model.add(MaxPooling2D(pool_size=(2, 2)))
```

### 3.3 Слой Dense (плотный слой)

**Dense Layer** в Keras - это **полносвязный слой**, который применяется в нейронных сетях для выполнения линейной операции над входными данными. 

Он принимает на вход одномерный тензор и производит линейное преобразование, умножая входные данные на **матрицу весов** и добавляя **смещение (bias)**. Затем, результат проходит через функцию активации, которая добавляет нелинейность в модель. 

image.png

Dense Layer может содержать разное количество нейронов и использовать различные функции активации, в зависимости от задачи. Например, для задачи бинарной классификации можно использовать слой с одним нейроном и функцией активации sigmoid, а для задачи многоклассовой классификации - слой с количеством нейронов, соответствующим числу классов, и функцией активации softmax.

```python
model.add(keras.layers.Dense(128, activation='sigmoid'))
model.add(keras.layers.Dense(10, activation='softmax'))
```

### 3.4 Слой Dropout (слой исключения)

**Dropout Layer** в Keras - это слой, используемый в нейронных сетях для борьбы с переобучением (используется в качестве метода **регуляризации** для предотвращения переобучения в нейронных сетях)

Он работает путем **случайного обнуления (отключения) некоторых нейронов** во время обучения. Каждый нейрон в слое имеет определенную вероятность быть обнуленным (обычно 0,5), которая задается в качестве гиперпараметра. 

В результате, некоторые нейроны не участвуют в процессе обучения на каждой итерации, что позволяет избежать переобучения и улучшить обобщающую способность модели. 

Во время тестирования, все нейроны активируются и выход слоя масштабируется на вероятность активации каждого нейрона. Dropout Layer может быть добавлен в любое место нейронной сети, где есть риск переобучения, например, между Dense Layers или Convolutional Layers.

![image](https://www.deeplearning.ai/_next/image/?url=https%3A%2F%2Fdl-staging-website.ghost.io%2Fcontent%2Fimages%2F2021%2F08%2FDropout-With-a-Difference-1.gif&w=2048&q=75)

Мы можем добавить слой отсева с вероятностью 50% следующим образом:

```python
model.add(Dropout(0.5))
```

### 3.5 Embedding layer (слой внедрения)

**Embedding Layer** в Keras - это слой, используемый в нейронных сетях для преобразования дискретных категориальных признаков, таких как слова или теги, в **плотные векторы низкой размерности**. 

Он работает путем создания отображения между каждым уникальным значением категориального признака и соответствующим вектором. Векторы оптимизируются в процессе обучения нейронной сети, чтобы минимизировать ошибку задачи, например, классификации текста или прогнозирования следующего слова. 

Embedding Layer позволяет представлять **категориальные признаки в виде непрерывных векторов**, что улучшает качество модели и упрощает ее обучение. Размерность выхода Embedding Layer определяется гиперпараметрами слоя, такими как размер словаря и размерность векторов.

![Image](https://mayankjain.ca/blog-posts/images/17-13.gif)

### 3.6 Flatten layer (слой внедрения)

**Flatten Layer** в Keras - это слой, используемый в нейронных сетях для преобразования многомерных тензоров в одномерный вектор. 

Он принимает на вход тензор произвольной формы и выравнивает его в одномерный вектор, сохраняя все элементы исходного тензора. Flatten Layer обычно используется для преобразования выхода сверточных слоев в одномерный вектор перед передачей его в полносвязные слои. Это позволяет использовать информацию о пространственной структуре изображения при обучении нейронной сети, а также уменьшить количество параметров в модели. 

Кроме того, Flatten Layer может использоваться для преобразования любых многомерных тензоров, например, для обработки последовательностей или временных рядов.

image.png

```python
model.add(Flatten())
```

## 4. Компиляция и обучение модели

### 4.1 Компиляция с помощью метода .compile()

После того, как мы определим нашу модель, мы обучаем ее. Сначала необходимо скомпилировать сеть с **функцией потерь** и **функцией оптимизатора**. Это позволит сети менять веса и минимизировать потери.

Мы скомпилируем нашу модель с помощью **метода .compile()** следующим образом:

```python
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

При компиляции есть возможность выбора функции потерь и оптимизатор:

**Функции потерь:**

- **Среднеквадратическая ошибка (MSE):** обычная функция потерь для задач регрессии — измеряет среднеквадратичную разницу между прогнозируемыми и фактическими значениями.

- **Двоичная кроссэнтропия:** функция потерь для задач бинарной классификации — измеряет кросс-энтропию между предсказанным и фактическим бинарным распределениями.

- **Категориальная кроссэнтропия:** функция потерь для задач классификации с несколькими классами — измеряет кросс-энтропию между прогнозируемым и фактическим категориальным распределением.

**Оптимизаторы:**

- **Стохастический градиентный спуск (SGD):** простой алгоритм оптимизации, который обновляет параметры путем вычисления градиента функции потерь по отношению к параметрам.

- **Adam:** алгоритм оптимизации, который адаптирует скорость обучения на основе исторической информации о градиенте.

- **RMSprop:** алгоритм оптимизации, который использует скользящее среднее квадратов градиентов для нормализации обновлений градиента.

### 5.1. Работа с числовыми данными

**Числовые значения** будут наиболее частыми типами данных, с которыми приходится иметь дело. Несмотря на то, что они уже в подходящем для расчетов формате, все же их подвергают дополнительным преобразованиям. Основная проблема с числовыми данными заключается в том, что каждый признак имеет **свой масштаб**. 

Рассмотрим набор данных о ценах на жилье с информацией о размере дома, количестве комнат, времени постройки и цене. Предположим, наша цель — спрогнозировать цену, учитывая размер дома, количество комнат и время постройки. 

Каждая из этих характеристик представлена ​​в разном масштабе. Размер дома может варьироваться, скажем, от 100 до 500 квадратных метров, год постройки — это четырехзначное число, которое может достигать 200-летней давности, количество комнат — это число от 1 до 4.

Проблема здесь в том, что модель может уделять больше внимания одному признаку в зависимости от его ценности. Таким образом, размер дома может привлечь больше внимания, поскольку его значения больше, а другими важными характеристиками, такими как количество комнат, можно пренебречь, поскольку их значения малы.

У нас есть два простых решения этой проблемы, они называются **нормализация** и **стандартизация**.

**Нормализация**

Нормализация просто масштабирует значения в диапазоне [0-1]. Чтобы применить его к набору данных, вам просто нужно вычесть минимальное значение из каждой функции и разделить его на диапазон (максимум — минимум).

**Стандартизация**

С другой стороны, стандартизация преобразует данные так, чтобы они имели нулевое среднее значение и стандартное отклонение, равное одной единице.

## 6. Keras в действии — пример простой линейной регрессии

```python
# Импорт необходимых модулей
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation
import numpy as np
import matplotlib.pyplot as plt

 
# Определение данных для модели
x = data = np.linspace(1, 2, 200)
y = x*4 + np.random.randn(*x.shape) * 0.3

# Создание последовательной (Sequential) модели
model = Sequential()

# Добавление слоев в Sequential model
model.add(Dense(1, input_dim=1, activation='linear'))

# Компиляция модели
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

# Объявление начальных весов и смещения
weights = model.layers[0].get_weights()
w_init = weights[0][0][0]
b_init = weights[1][0]
print('Модель линейной регрессии инициализируется с весами w: %.2f, b: %.2f' % (w_init, b_init)) 

# Обучение модели
model.fit(x,y, batch_size=1, epochs=30, shuffle=False)

# Установка окончательных весов и смещения
weights = model.layers[0].get_weights()
w_final = weights[0][0][0]
b_final = weights[1][0]
print('Модель линейной регрессии обучена c весом w: %.2f, b: %.2f' % (w_final, b_final))

# Предсказания результатов
predict = model.predict(data)

# Visualize the results
plt.figure(figsize=(12,8))
plt.plot(data, predict, 'b', data , y, 'k.')
plt.show()
```

После обучения данных выходные данные должны выглядеть, как на приведенном выше графике.

```python
from keras import models, layers
model = models.Sequential()
model.add(layers.Dense(512, activation='relu', input_shape=(28*28, )))
model.add(layers.Dense(10, activation='softmax'))
model.compile('rmsprop', 'categorical_crossentropy', metrics=['acc'])
```

Здесь у нас есть две новые вещи: 

1) активация = 'softmax' 

2) потери = 'categorical_crossentropy'.

**Softmax** — это специальная функция активации, которая преобразует выходные данные в значения вероятности каждого класса. Следовательно, с **Dense(10)** у нас будет 10 нейронов, каждый из которых представляет вероятность данной цифры.

**Категориальная кросс-энтропия** — это просто функция потерь, которая вычисляет ошибку между предсказанной цифрой и фактической. Мы используем **категориальную кросс-энтропию**, когда у нас есть 3 или более классов, и **бинарную кросс-энтропию**, когда у нас есть 2 класса.
