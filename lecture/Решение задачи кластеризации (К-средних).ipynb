Алгоритм следует простому или простому способу классификации заданного набора данных с помощью определенного количества кластеров, фиксированных априори. Алгоритм K-средних работает итеративно, чтобы назначить каждую точку данных одной из групп K на основе предоставленных функций. Точки данных группируются на основе сходства признаков.

# 3. Принцип кластеризации K-средних

Кластеризация K-средних используется для поиска внутренних групп в немаркированном наборе данных и извлечения из них выводов. Он основан на кластеризации на основе `центроида`.

`Центроид` — это точка данных в центре кластера. Это итеративный алгоритм, в котором понятие сходства определяется тем, насколько близко точка данных находится к центроиду кластера. 

Кластеризация K-средних работает следующим образом: `алгоритм кластеризации K-средних использует итеративную процедуру для получения окончательного результата. Алгоритму требуется количество кластеров K и набор данных в качестве входных данных.` 


![Clustering](https://upload.wikimedia.org/wikipedia/commons/e/ea/K-means_convergence.gif?20170530143526)

Набор данных представляет собой набор функций для каждой точки данных. Алгоритм начинается с начальных оценок для `K центроидов`. Затем алгоритм выполняет итерацию между двумя этапами:

### 3.1 Шаг присвоения данных

Каждый центроид определяет один из кластеров. На этом этапе каждой точке данных присваивается ее ближайший центроид, который основан на квадрате евклидова расстояния. Итак, если $c_j$ представляет собой набор центроидов в наборе C, то каждая точка данных назначается кластеру на основе минимального евклидова расстояния.

![image](https://www.saedsayad.com/images/Clustering_kmeans_c.png)

### 3.2 Шаг обновления центроида

На этом этапе центроиды пересчитываются и обновляются. Это делается путем получения среднего значения всех точек данных, назначенных кластеру этого центроида.

Затем алгоритм выполняет итерацию между шагами 1 и 2 до тех пор, пока не будет достигнут критерий остановки. 

`Критерий остановки` означает, что никакие точки данных не изменяют кластеры, сумма расстояний минимизируется или достигается некоторое максимальное количество итераций. Этот алгоритм гарантированно сходится к результату. Результатом может быть локальный оптимум, означающий, что оценка более чем одного прогона алгоритма с рандомизированными начальными центроидами может дать лучший результат.

# 4. Выбор значения К (число кластеров)

Алгоритм K-Means зависит от нахождения количества кластеров и меток данных для `заранее определенного значения K`. Чтобы найти количество кластеров в данных, нам нужно запустить алгоритм кластеризации K-Means для различных значений K и сравнить результаты. Таким образом, производительность алгоритма K-Means зависит от значения K. 

Мы должны выбрать оптимальное значение K, которое дает нам наилучшую производительность. Существуют различные методы для нахождения оптимального значения K. Наиболее распространенным методом является `метод локтя`.

![image](https://av-eks-blogoptimized.s3.amazonaws.com/62725cluster0.png)

# 5. Метод локтя

Метод локтя используется для определения оптимального количества кластеров в кластеризации K-средних. Метод локтя строит `значение функции стоимости`, созданное различными значениями K. 

Если K увеличивается, `среднее искажение (суммарное значение расстояний на инертность)` уменьшается. Тогда в каждом кластере будет меньше составных экземпляров, и экземпляры будут ближе к своим соответствующим центроидам. Однако улучшение среднего искажения будет снижаться по мере увеличения K. 

Значение K, при котором улучшение искажения снижается больше всего, называется `локтем`, при котором `мы должны прекратить разделение данных на дополнительные кластеры`.

- Алгоритм KMeans группирует данные, пытаясь разделить выборки на n групп с равными отклонениями, минимизируя критерий, известный как `инерция`, или `критерий суммы квадратов внутри кластера`. Инерция или критерий суммы квадратов внутри кластера может быть распознан как мера того, насколько `внутренне когерентны кластеры`.
  
- Алгоритм k-средних делит набор из N выборок X на K непересекающихся кластеров C, каждый из которых описывается средним значением j выборок в кластере. Средние значения обычно называют центроидами кластера.
  
- Алгоритм K-средних направлен на выбор центроидов, которые минимизируют инерцию или сумму квадратов критерия внутри кластера.

# Инерция

- Инерция не является нормированной метрикой.

- Более низкие значения инерции лучше, а ноль оптимален.

- Но в очень многомерных пространствах евклидовы расстояния имеют тенденцию к завышению (это пример curse of dimensionality).

- Запуск алгоритма уменьшения размерности, такого как PCA, перед кластеризацией k-средних может облегчить эту проблему и ускорить вычисления.

Мы можем рассчитать инерцию модели следующим образом:

```python
kmeans.inertia_
```

- Чем меньше инерция модели, тем лучше модель подходит.

- Мы видим, что модель имеет очень высокую инерцию. Таким образом, это не очень хорошая модель, подходящая для данных.

# 15. Спрогнозируем оптимальное количество кластеров

```python
from sklearn.cluster import KMeans
cs = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    cs.append(kmeans.inertia_)
plt.plot(range(1, 11), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('CS')
plt.show()
```

- На приведенном выше графике мы видим, что при k = 2 есть излом.

- Следовательно, k=2 можно считать хорошим числом кластера для кластеризации этих данных.

- Но мы видели, что я добился слабой точности в 1% при k=2.

Снова повторим код с k=2 для удобства.

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2,random_state=0)
kmeans.fit(X)
labels = kmeans.labels_

correct_labels = sum(y == labels)
print("Результат: %d из %d образцов были помечены правильно." % (correct_labels, y.size))
print('Оценка точности: {0:0.2f}'. format(correct_labels/float(y.size)))
```

Таким образом, наша неконтролируемая модель кластеризации достигла очень слабой точности в 1%.

### Визуальный анализ

```python
X1 = X[['num_reactions', 'status_type']].iloc[: , :].values
inertia = []
for n in range(1 , 11):
    algorithm = (KMeans(n_clusters = n, init='k-means++', n_init = 10, max_iter=300, 
                        tol=0.0001,  random_state= 111  , algorithm='elkan') )
    algorithm.fit(X1)
    inertia.append(algorithm.inertia_)
```

```python
plt.figure(1 , figsize = (15 ,6))
plt.plot(np.arange(1 , 11) , inertia , 'o')
plt.plot(np.arange(1 , 11) , inertia , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('Inertia')
plt.show()
```

```python
algorithm = (KMeans(n_clusters = 4 ,init='k-means++', n_init = 10 ,max_iter=300, tol=0.0001,  random_state= 111  , algorithm='elkan') )
algorithm.fit(X1)
labels1 = algorithm.labels_
centroids1 = algorithm.cluster_centers_
```

```python
h = 0.02
x_min, x_max = X1[:, 0].min() - 1, X1[:, 0].max() + 1
y_min, y_max = X1[:, 1].min() - 1, X1[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = algorithm.predict(np.c_[xx.ravel(), yy.ravel()]) 
```

```python
plt.figure(1 , figsize = (15 , 7) )
plt.clf()
Z = Z.reshape(xx.shape)
plt.imshow(Z , interpolation='nearest', extent=(xx.min(), xx.max(), yy.min(), yy.max()), cmap = plt.cm.Pastel2, aspect = 'auto', origin='lower')
plt.scatter(x = X1[: , 0], y = X1[: , 1] , c = labels1, s=10, alpha = 0.5)
plt.scatter(x = centroids1[: , 0] , y =  centroids1[: , 1] , s = 300 , c = 'red' , alpha = 0.5)
plt.ylabel('num_reactions') , plt.xlabel('status_type')
plt.show()
```
