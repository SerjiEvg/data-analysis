## Преимущества NumPy

Вот четыре основных преимущества, которые дает при использовании NumPy:

 - `Больше скорости`: NumPy использует алгоритмы, написанные на C, которые выполняются за наносекунды, а не секунды.

- `Меньше циклов`: NumPy помогает сократить количество циклов и не запутаться в индексах итераций.

- `Более понятный код`: без циклов код будет больше похож на уравнения, которые просто вычисляются.

- `Качество кода`: тысячи участников сообщества работают над тем, чтобы NumPy работал быстро, удобно и без ошибок.

Из-за этих преимуществ NumPy является `стандартом де-факто для многомерных массивов` в науке о данных Python, и многие из самых популярных библиотек созданы на его основе. 

Изучение NumPy — отличный способ заложить прочную основу, и потом расширять свои знания в более конкретных областях науки о данных.

## Форма и атрибуты массива

Все массивы имеют форму, доступную с помощью `.shape`. Например, давайте получим форму вектора, матрицы и тензора.

```python
vector = np.arange(5)
print("Vector shape", vector.shape)
```

```python
matrix = np.ones([3, 2])
print("Matrix shape", matrix.shape)
```

```python
tensor = np.zeros([2, 3, 3])
print("Tensor shape", tensor.shape)
```

Форма вектора одномерная. Первое число в его форме — это количество элементов (или строк). 

Для матрицы `.shape` говорит нам, что у нас есть три строки и два столбца. 

Тензор немного другой. Первое число — это количество матриц/срезов, которые у нас есть. Второй дает количество строк. Третий обеспечивает количество столбцов.

```python
import numpy as np
np.random.seed(0)  # воспроизводимость

x1 = np.random.randint(-10, 10, size=6)  # 1-мерный массив
x2 = np.random.randint(10, size=(3, 4))  # 2-мерный массив
x3 = np.random.randint(10, size=(3, 4, 5))  # 3-мерный массив
x2
```

Полезным атрибутом является `dtype`, тип данных массива

```python
print("dtype:", x3.dtype)
```

Другие атрибуты включают `itemsize`, в котором указан размер (в байтах) каждого элемента массива, и `nbytes`, в котором указан общий размер (в байтах) массива:

```python
print("itemsize (размер элемента):", x3.itemsize, "bytes")
print("nbytes (объем массива):", x3.nbytes, "bytes")
```

## Способы заполнения массивов

Библиотека numpy предоставляет несколько удобных функций для создания массивов желаемых форм с фиксированными или случайными значениями, помимо тех, которые мы уже использовали.

### Заполнение массива фиксированными значениями

Для того, чтобы все элементы массива были фиксированы, то есть одинаковыми, но отличными от 0 или 1, то можно использовать функцию `full(shape, fill_value)`, которая принимает минимум два параметра `shape` - размерность массива и `fill_value` - значение, которыми должен быть заполнен массив

```python
# Фиксированные значения
np.full([2, 3], 42)
```

### Заполнение массива значениями из диапазона

Можно заполнить массив значениями в указанном интервале и с определенным шагом. Для таких целей можно использовать функцию `arange(start, stop, step)`, где `start` — начало интервала, `stop` — конец интервала, `step` — интервал между значениями

```python
# Заполнение из диапазона (начало, конец, шаг)
np.arange(10, 90, 3)
```

### Заполнение массива по прогрессии

Иногда может потребоваться сгенерировать определенное количество данных в указанном интервале и одинаковым шагом. Для этих целей используется команда `linspace(start, stop, num, endpoint)`, которая может принимать 4 параметра (два первые из которых являются обязательными), где

- `start`, `stop` - начало и конец отсчета,
- `num` - количество данных в выборке, 
- `endpoint` — определение конца выборки (если True (по умолчанию), то выборка заканчивается на последнем значении, если False — последнее значение в выборку не включается).

```python
# Заполнение значениями с указанным шагом
np.linspace(3, 27, 9)
```

# 3. Индексация в массивах NumPy

В какой-то момент возникнет необходимость индексировать (выбирать) подмножества массива NumPy. 

Например, необходимо выбрать один столбец данных или выполнить манипуляцию с этим столбцом. Нотация индексации в NumPy абсолютно такая же, как и в Python.

 В `индексировании массивов` используются многие из тех же идиом, что и в обычном коде Python. Мы можем использовать положительные или отрицательные индексы для индексации с начала или конца массива. Мы можем использовать двоеточие (:), чтобы указать «остальное» или «все». Можно использовать даже два двоеточия, чтобы пропустить элементы, как в обычных списках Python.

`Но есть небольшая разница: в массивах NumPy между осями используются запятые, поэтому мы можем индексировать несколько осей в одном наборе квадратных скобок. `

## Обозначения индексации

- Запятые разделяют оси массива.

- Двоеточие означает «через». Например, x[0:4] означает первые 5 строк (строки с 0 по 4) вектора x.

- Отрицательные числа означают «с конца массива». Например, x[-1] означает последнее значение вектора x.

- Пробелы перед двоеточием или после него означают «остальное». Например, x[3:] ​​означает остальные значения в x после 3-го элемента. Аналогично, x[:3] означает все значения до 3-го элемента. x[:] означает все значения вектора x.

- Когда индексов меньше, чем осей, отсутствующие индексы считаются полными срезами. Например, в 3-осевом массиве x[0, 0] означает все данные по 3-й оси 1-й строки и 1-го столбца.

- Точки "..." означают столько двоеточий, сколько необходимо для создания полного кортежа индексации. Например, x[1, 2, ...] совпадает с x[1, 2, :, :, :].

## Нарезка массивов

Numpy интуитивно расширяет нотацию индексации списка Python, используя несколько измерений [ ] (по сути мы увеличиваем количество размерностей). Мы можем предоставить список индексов или диапазонов, разделенных запятыми, чтобы выбрать определенный элемент или подмассив (также называемый срезом) из массива numpy.

Например дан массив `arr1`, который имеет размерность (3, 2, 4)

```python
arr1 = np.array([
    [
        [11, 12, 13, 14], 
        [13, 14, 15, 19]
    ], 
    
    [
        [15, 16, 17, 21], 
        [63, 92, 36, 18]
    ], 
    
    [
        [98, 32, 81, 23],      
        [17, 18, 19.5, 43]
    ]
])

arr1
```

Как и в случае со списками, мы можем обратиться к элементу по индексу. В нашем случае, массив `arr1` трехмерный, поэтому если нужно получить конкретное число из массива, то необходимо прописать 3 индекса

```python
# Получаем один элемент
arr1[1, 1, 2]
```

Кроме обращения по индексу, мы можем получить срез из массива.

Например: 

```python
# Срез массива с использованием диапазона
arr1[1:, 0:1, :2]
```

также мы можем комбинировать индексы, например, в качестве первого индекса мы можем указать срез, а в качестве других числа:

## Индексация (фильтрация) по маске

Другой удобный способ проиндексировать определенные разделы массива NumPy — использовать массив масок. 

`Маска` — это массив, который имеет ту же форму, но вместо наших значений содержит логические значения: `True` или `False`.  Индексация данного элемента массива определяется значением соответствующего элемента массива масок. Этот массив масок используется для индексации массива данных нелинейным и сложным образом. Он возвращает все те элементы, где логический массив имеет значение True.

Во-первых, мы определяем массив значений True/False NumPy, где значения True — это те, которые мы хотим сохранить. Затем мы маскируем массив, например, `subset`. В результате сохраняются только строки, соответствующие элементам, которые имеют значение True в массиве маски.

```python
import random
subset = np.array([random.random() for i in range(10)])
subset
```

```python
mask_array = np.array([False, True, True, False, True, True, False, True, False, True])

subset[mask_array]
```

Как видите, массив масок сохранил строки, соответствующие True , и исключил строки, соответствующие False. 

`Маскирование` — это мощный инструмент, который позволяет нам индексировать элементы на основе логических выражений. 

 Рассмотрим следующий пример:

```python
import numpy as np

numbers = np.linspace(5, 50, 24, dtype=int).reshape(4, -1)

numbers
```

Следующая операция создает маску, выполняя `векторизованное логическое вычисление`. То есть берется каждый элемент и проверяется - делится ли он на четыре без остатка. На основании этого возвращает массив масок той же формы с поэлементными результатами вычисления.

```python
mask = numbers % 4 == 0

mask
```

Эту маску мы используем для индексации исходного массива numbers. Массив теряет свою первоначальную форму, уменьшаясь до одного измерения, но мы все равно получаем необходимые данные.

```python
numbers[mask]
```

Следующая операция, обеспечивает более традиционный, идиоматический маскированный выбор.

```python
by_four = numbers[numbers % 4 == 0]

by_four
```

# 4. Обработка массивов NumPy

## Объединение

NumPy также предоставляет полезные функции для `конкатенации` (т.е. соединения) массивов.  Хотя имеется функция `np.concatenate()`, есть также ряд вспомогательных функций , которые иногда легче читать.

```python
a = np.array([
    [4, 8],
    [6, 1]
])

b = np.array([
    [3, 5],
    [7, 2]
])

np.hstack((a, b))
```

Все эти функции принимают в качестве первого аргумента кортеж массивов, а не переменное число аргументов, как можно было бы ожидать. 

```python
np.vstack((b, a))
```

```python
np.concatenate((a, b))
```

```python
np.concatenate((a, b), axis=None)
```

## Разделение
Противоположностью конкатенации (то есть объединению) массивов является их **разделение**. Чтобы разделить массив, NumPy предоставляет следующие команды:

- `hsplit`: массив разбивается по горизонтальной оси.

- `vsplit`: массив разбивается по вертикальной оси.

- `dsplit`: массив разбивается по 3-й оси (глубина).

- `array_split`: позволяет указать ось для использования при разделении.

## Добавление/удаление элементов
NumPy предоставляет несколько функций для добавления или удаления данных из массива:

- `resize`: возвращает новый массив указанной формы с нулями в качестве заполнителей во всех новых ячейках.

- `append`: добавляет значения в конец массива.

- `insert`: добавляет значения в середину массива.

- `delete`: возвращает новый массив с удаленными данными.

- `unique`: находит только уникальные значения массива.

 ## Создание копий массивов

Несмотря на приятные особенности представлений массивов, иногда бывает полезно вместо этого явно копировать данные внутри массива или подмассива. Проще всего это сделать с помощью метода `copy()`:

```python
import numpy as np
np.random.seed(0)  # обновление  
x2 = np.random.randint(10, size=(3, 4)) 

x2_sub_copy = x2[:2, :2].copy()
print(x2_sub_copy)
```

Если мы сейчас изменим этот подмассив, исходный массив не будет затронут:

```python
x2_sub_copy[0, 0] = 42
print(x2_sub_copy)
```

```python
print(x2)
```

 ## Изменение формы массивов

Другой полезный тип операции — `изменение формы массивов`. Наиболее гибкий способ сделать это с помощью  метода `reshape`. Например, если вы хотите поместить числа от 1 до 9 в 3×3 сетки, вы можете сделать следующее:

```python
grid = np.arange(1, 10).reshape((3, 3))
print(grid)
```

Обратите внимание, что для того, чтобы это работало, размер исходного массива должен соответствовать размеру измененного массива. Там, где это возможно, reshape метод будет использовать представление исходного массива без копирования, но с несмежными буферами памяти это не всегда так.

Другим распространенным шаблоном изменения формы является `преобразование одномерного массива в двумерную матрицу строк или столбцов`. Это можно сделать с помощью метода `reshape` или проще сделать с помощью ключевого  слова `newaxis` в операции среза:

```python
x = np.array([1, 2, 3])

# в строку через reshape
x.reshape((1, 3))
```

```python
# в стороку через newaxis
x[np.newaxis, :]
```

```python
# в столбец через reshape
x.reshape((3, 1))
```

```python
# в столбец через newaxis
x[:, np.newaxis]
```

 ## Векторизованные функции

NumPy предоставляет версии стандартных функций `log`, `exp`, `sin` и т.д., которые действуют на массивы поэлементно.

```python
z = np.array([1, 2, 3])
np.sin(z)
```

Это устраняет необходимость в явных поэлементных циклах, таких как

```python
n = len(z)
y = np.empty(n)
for i in range(n):
    y[i] = np.sin(z[i])
```

Поскольку они действуют на массивы поэлементно, эти функции называются `векторизованными функциями`.

На языке NumPy их также называют `ufuncs`.

Как мы видели выше, обычные арифметические операции ( +, *, и т. д.) также работают поэлементно, и их объединение с `ufuncs` дает очень большой набор быстрых поэлементных функций.

```python
z
```

```python
(1 / np.sqrt(2 * np.pi)) * np.exp(- 0.5 * z**2)
```

Не все пользовательские функции будут действовать поэлементно.

Например, передача функции f, вызывает `ValueError`

```python
def f(x):
    return 1 if x > 0 else 0
```

Поэтому поступают так:

```python
x = np.random.randn(4)
x
```

```python
np.where(x > 0, 1, 0)  # Insert 1 if x > 0 true, otherwise 0
```

```python
f = np.vectorize(f)
f(x)  
```

Однако этот подход не всегда обеспечивает ту же скорость, что и более тщательно разработанная векторизованная функция.

 ## Сравнения

```python
z = np.array([2, 3])
y = np.array([2, 3])
z == y
```

## Подпакеты

NumPy предоставляет некоторые дополнительные функции, связанные с научным программированием, через свои подпакеты.

Мы уже видели, как мы можем генерировать случайные величины, используя `np.random`.

```python
z = np.random.randn(10000)  # Generate standard normals
y = np.random.binomial(10, 0.5, size=1000)    # 1,000 draws from Bin(10, 0.5)
y.mean()
```

Другим часто используемым подпакетом является `np.linalg`.

```python
A = np.array([[1, 2], [3, 4]])

np.linalg.det(A)           # Вычисляем определитель
```

```python
np.linalg.inv(A)           # Вычисляем обратную матрицу
```

Большая часть этой функциональности также доступна в SciPy, наборе модулей, построенных на основе NumPy.
