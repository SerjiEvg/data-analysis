# Введение в анализ данных

**Анализ данных** — это процесс очистки, изменения и обработки необработанных данных, а также извлечения полезной и актуальной информации, которая помогает предприятиям принимать обоснованные решения. 

Данные представленные в количественном виде, часто называют **информацией**. Анализ данных представляет собой процесс получения информации путем создания моделей и применения математического аппарата (точнее статистических законов) для поиска закономерностей.

Процедура помогает снизить риски, связанные с принятием решений, предоставляя полезную информацию и статистику, часто представленную в виде диаграмм, изображений, таблиц и графиков.

**Наука о данных** - это процесс использования статистики и анализа данных для понимания явлений, скрытых в данных. В науке о данных проводится сложный анализ на основе знаний из разных предметных областей: математики, статистики, информатики, машинного обучения, баз данных и т.д.

Данные поступают в различных структурах , форматах и ​​типах, включая следующие:

- **Большие данные**. Большие данные определяются  как огромный набор данных, который со временем продолжает расти в геометрической прогрессии. Четыре основные  характеристики больших данных  : объем, разнообразие, скорость и изменчивость. Объем описывает количество, скорость относится к скорости роста данных, а разнообразие указывает на разные источники данных. Достоверность говорит о качестве данных , определяя, представляют ли они ценность для бизнеса или нет.

- **Структурированные/неструктурированные данные**. Структурированные данные — это предопределенная модель данных, такая как традиционная база данных со столбцами строк. Неструктурированные данные поступают в формате, который не помещается в строки и столбцы, и могут включать в себя видео, фотографии, аудио, текст и многое другое. Сравнение  структурированных данных с неструктурированными  показывает, что структурированными данными легче управлять и анализировать.

- **Метаданные**. Метаданные  — это форма данных, которая описывает и предоставляет информацию о других данных. Например, метаданные для изображения могут включать автора, тип изображения и дату создания. Метаданные позволяют пользователям упорядочивать неструктурированные данные по категориям, что упрощает работу с ними.

- **Данные в режиме реального времени**. Данные, которые представляются, как только они получены, известны как  данные в реальном времени. Этот тип данных полезен, когда решения требуют самой последней информации. Например, биржевой маклер может использовать биржевой тикер для отслеживания  наиболее активных акций в режиме реального времени.

- **Машинные данные**. Благодаря Интернету вещей (IoT), датчикам и другим технологиям данные могут автоматически генерироваться заводскими системами и другими машинами, информационными технологиями и телекоммуникационной инфраструктурой, умными автомобилями, портативными устройствами и многим другим. Этот тип данных известен как  машинные данные , потому что он полностью создается машинами без участия человека.


Методы и методы анализа данных полезны для поиска информации в данных, таких как показатели, факты и цифры. Двумя основными методами анализа данных являются методы качественного анализа данных и методы количественного анализа данных. Эти методы анализа данных можно использовать независимо или в сочетании с другими, чтобы помочь бизнес-руководителям и лицам, принимающим решения, получать информацию о бизнесе из различных типов данных .

## Процесс анализа данных

Ответ на вопрос «что такое анализ данных» — это только первый шаг. Теперь посмотрим, как это выполняется. Процесс анализа данных или, альтернативно, этапы анализа данных включают в себя сбор всей информации, ее обработку, изучение данных и их использование для поиска закономерностей и других идей. Процесс состоит из:

- **Сбор требований к данным:** спросите себя, почему вы проводите этот анализ, какой тип анализа данных вы хотите использовать и какие данные вы планируете анализировать.

- **Сбор данных:** руководствуясь определенными вами требованиями, пришло время собирать данные из ваших источников. Источники включают тематические исследования, опросы, интервью, анкеты, прямое наблюдение и фокус-группы. Обязательно систематизируйте собранные данные для анализа.

- **Очистка данных:** не все данные, которые вы собираете, будут полезны, поэтому пришло время их очистить. В этом процессе вы удаляете пробелы, повторяющиеся записи и основные ошибки. Очистка данных обязательна перед отправкой информации на анализ.

- **Анализ данных:** здесь вы используете программное обеспечение для анализа данных и другие инструменты, которые помогут вам интерпретировать и понимать данные и делать выводы. Инструменты анализа данных включают Excel, Python , R, Looker, Rapid Miner, Chartio, Metabase, Redash и Microsoft Power BI.

- **Интерпретация данных:** теперь, когда у вас есть результаты, вам нужно их интерпретировать и предложить наилучший план действий на основе ваших выводов.

- **Визуализация данных:** визуализация данных — это причудливый способ сказать: «графически покажите свою информацию так, чтобы люди могли ее прочитать и понять». Вы можете использовать диаграммы, графики, карты, маркеры или множество других методов. Визуализация помогает получать ценные сведения, помогая сравнивать наборы данных и наблюдать за взаимосвязями.


## Виды анализа данных

Сегодня доступно полдюжины популярных типов анализа данных, обычно используемых в мире технологий и бизнеса. Они есть: 

- **Диагностический анализ:** диагностический анализ отвечает на вопрос: «Почему это произошло?» Используя информацию, полученную в результате статистического анализа (подробнее об этом позже!), аналитики используют диагностический анализ для выявления закономерностей в данных. В идеале аналитики находят аналогичные модели, существовавшие в прошлом, и, следовательно, используют эти решения для решения текущих проблем.

- **Предиктивный анализ:** предиктивный анализ отвечает на вопрос: «Что, скорее всего, произойдет?» Используя закономерности, обнаруженные в старых данных, а также в текущих событиях, аналитики предсказывают будущие события. Хотя не существует такой вещи, как 100-процентно точный прогноз, шансы улучшаются, если у аналитиков есть много подробной информации и дисциплина для ее тщательного исследования.

- **Предписывающий анализ:** cмешайте все выводы, полученные из других типов анализа данных, и вы получите предписывающий анализ. Иногда проблема не может быть решена только с помощью одного типа анализа, вместо этого требуется несколько аналитических данных.

- **Статистический анализ:** cтатистический анализ отвечает на вопрос: «Что произошло?» Этот анализ охватывает сбор данных, анализ, моделирование, интерпретацию и представление с использованием информационных панелей. Статистический анализ делится на две подкатегории:

- **Описательный:** описательный анализ работает либо с полными, либо с выборками обобщенных числовых данных. Он иллюстрирует средние значения и отклонения в непрерывных данных, а также проценты и частоты в категориальных данных.

- **Инференциальный:** инференциальный анализ работает с образцами, полученными из полных данных. Аналитик может прийти к разным выводам из одного и того же обширного набора данных, просто выбрав разные выборки.

- **Анализ текста:** также называемый « интеллектуальным анализом данных », анализ текста использует базы данных и инструменты интеллектуального анализа данных для обнаружения шаблонов, находящихся в больших наборах данных. Он преобразует необработанные данные в полезную бизнес-информацию. Анализ текста, пожалуй, самый простой и прямой метод анализа данных.

## Анализ данных в исследованиях

Исследователи в значительной степени полагаются на данные, поскольку им есть что рассказать или решить проблемы. Все начинается с вопроса, а данные — не что иное, как ответ на этот вопрос. Но что, если нет вопроса, чтобы задать? 

Данные можно исследовать даже без проблем — мы называем это «интеллектуальным анализом данных», который часто выявляет некоторые интересные закономерности в данных, которые стоит изучить.

Независимо от типа данных, которые исследуют исследователи, их миссия и видение аудитории помогают им найти закономерности, формирующие историю, которую они хотят рассказать. Одна из важнейших вещей, которую ожидают от исследователей при анализе данных, — оставаться открытыми и оставаться непредвзятыми к неожиданным закономерностям, выражениям и результатам. 

Помните, иногда анализ данных рассказывает самые непредвиденные, но захватывающие истории, которые не ожидались во время начала анализа данных. Поэтому полагайтесь на имеющиеся у вас данные и наслаждайтесь поисковым исследованием. 

### Типы данных в исследованиях

Каждый вид данных обладает редким качеством описания вещей после присвоения им определенного значения. Для анализа вам необходимо организовать эти значения, обработанные и представленные в заданном контексте, чтобы сделать их полезными. Данные могут быть в разных формах; вот основные типы данных.

- **Качественные данные:** когда представленные данные содержат слова и описания, мы называем это качественными данными . Хотя вы можете наблюдать за этими данными, они субъективны и их сложнее анализировать в исследованиях, особенно для сравнения. Пример: данные о качестве представляют собой все, что описывает вкус, опыт, текстуру или мнение, которые считаются данными о качестве. Этот тип данных обычно собирается с помощью фокус-групп, личных качественных интервью или с использованием открытых вопросов в опросах.

- **Количественные данные:** любые данные, выраженные числовыми цифрами, называются количественными данными . Этот тип данных можно разделить на категории, сгруппировать, измерить, рассчитать или ранжировать. Пример: такие вопросы, как возраст, ранг, стоимость, длина, вес, баллы и т. д. все подпадает под этот тип данных. Вы можете представить такие данные в графическом формате, диаграммах или применить к этим данным методы статистического анализа. Анкеты OMS (Системы измерения результатов) в опросах являются важным источником сбора числовых данных.

- **Категориальные данные:** это данные, представленные в группах. Однако элемент, включенный в категориальные данные, не может принадлежать более чем к одной группе. Пример: человек, отвечая на опрос, рассказывая о своем образе жизни, семейном положении, привычке курить или пить, попадает под категориальные данные. Тест хи-квадрат является стандартным методом, используемым для анализа этих данных.

### Поиск закономерностей в качественных данных

Хотя существует несколько способов поиска закономерностей в текстовой информации, метод на основе слов является наиболее надежным и широко используемым глобальным методом для исследований и анализа данных. Примечательно, что процесс анализа данных в качественных исследованиях выполняется вручную. Здесь исследователи обычно читают доступные данные и находят повторяющиеся или часто используемые слова. 

Например, изучая данные, собранные в африканских странах, чтобы понять самые насущные проблемы, с которыми сталкиваются люди, исследователи могут обнаружить, что  «еда»  и  «голод» являются наиболее часто используемыми словами, и выделят их для дальнейшего анализа.

Контекст ключевого слова — еще один широко используемый словесный метод. В этом методе исследователь пытается понять концепцию, анализируя контекст, в котором участники используют определенное ключевое слово.  

Например , исследователи, проводящие исследования и анализ данных для изучения понятия «диабет» среди респондентов, могут анализировать контекст того, когда и как респондент использовал или упоминал слово «диабет».

Техника на основе проверки также является одним из настоятельно рекомендуемых  методов анализа текста  , используемых для выявления шаблона качественных данных. Сравнение и противопоставление - широко используемый метод этой техники, позволяющий различать, насколько конкретный текст похож или отличается друг от друга. 

Например: Чтобы выяснить «важность штатного врача в компании», собранные данные делятся на людей, считающих необходимым нанимать штатного врача, и тех, кто считает это ненужным. Сравнить и сопоставить — лучший метод, который можно использовать для анализа опросов с типами вопросов с одним ответом.

Метафоры можно использовать, чтобы уменьшить кучу данных и найти в ней закономерности, чтобы стало легче связывать данные с теорией.

Разделение переменных — это еще один метод, используемый для разделения переменных, чтобы исследователи могли найти более связные описания и объяснения из огромного количества данных.

### Методы, используемые для анализа данных в количественных исследованиях

После того, как данные подготовлены для анализа, исследователи могут использовать различные методы исследований и анализа данных для получения осмысленных выводов. Безусловно, статистические методы являются наиболее предпочтительными для анализа числовых данных. Метод снова делится на две группы. Во- первых, «Описательная статистика» использовалась для описания данных. Во- вторых, «выводная статистика» , которая помогает сравнивать данные .

Описательная статистика
Этот метод используется для описания основных свойств разнородных типов данных в исследованиях. Он представляет данные таким осмысленным образом, что закономерности в данных начинают обретать смысл. Тем не менее описательный анализ не выходит за рамки выводов. Выводы снова основаны на гипотезе, сформулированной исследователями до сих пор. Вот несколько основных типов методов описательного анализа.

Меры частоты
- Количество, процент, частота
- Он используется для обозначения дома, часто происходит определенное событие.
- Исследователи используют его, когда хотят продемонстрировать, как часто дается ответ.
Меры центральной тенденции
- Среднее, медиана, мода
- Метод широко используется для демонстрации распределения по различным точкам.
- Исследователи используют этот метод, когда хотят продемонстрировать наиболее распространенный или усредненный ответ.
Меры дисперсии или вариации
- Диапазон, дисперсия, стандартное отклонение
- Здесь поле равняется высоким/низким точкам.
- Стандартное отклонение дисперсии = разница между наблюдаемой оценкой и средним значением
- Он используется для определения разброса оценок путем указания интервалов.
- Исследователи используют этот метод для демонстрации разбросанных данных. Это помогает им определить глубину распространения данных, что напрямую влияет на среднее значение.
Меры положения
- Процентильные ранги, Квартильные ранги
- Он опирается на стандартизированные оценки, помогающие исследователям определить взаимосвязь между различными оценками.
- Он часто используется, когда исследователи хотят сравнить баллы со средним значением.

Метафоры можно использовать, чтобы уменьшить кучу данных и найти в ней закономерности, чтобы стало легче связывать данные с теорией.

Разделение переменных — это еще один метод, используемый для разделения переменных, чтобы исследователи могли найти более связные описания и объяснения из огромного количества данных.

#### Методы, используемые для анализа данных в качественных исследованиях

Существует несколько методов анализа данных в качественных исследованиях, но вот некоторые из наиболее часто используемых:

- Контент-анализ:  это широко распространенный и наиболее часто используемый метод анализа данных в исследовательской методологии. Его можно использовать для анализа документированной информации из текста, изображений, а иногда и из физических предметов. Это зависит от вопросов исследования , чтобы предсказать, когда и где использовать этот метод.

- Нарративный анализ: этот метод используется для анализа контента, собранного из различных источников, таких как личные интервью, полевые наблюдения и  опросы . В большинстве случаев истории или мнения, которыми делятся люди, сосредоточены на поиске ответов на вопросы исследования.

- Анализ дискурса:  Подобно нарративному анализу, анализ дискурса используется для анализа взаимодействия с людьми. Тем не менее, именно этот метод учитывает социальный контекст, в котором или в рамках которого происходит общение между исследователем и респондентом. В дополнение к этому, анализ дискурса также фокусируется на образе жизни и повседневной среде, делая какие-либо выводы.

- Обоснованная теория:  если вы хотите объяснить, почему произошло то или иное явление, лучше всего использовать обоснованную теорию для анализа качественных данных. Обоснованная теория применяется для изучения данных о множестве подобных случаев, происходящих в разных условиях. Когда исследователи используют этот метод, они могут изменять объяснения или давать новые, пока не придут к какому-то заключению.

## Пакеты Python для науки о данных

Чтобы провести анализ данных в Python за минимальное время, мы должны сначала познакомиться с основными пакетами (библиотеки). 

Библиотека Python - это набор функций и методы, которые позволяют выполнять множество действий без написания кода. Библиотеки обычно содержат встроенные модули, содержащие разные функции, которые можно использовать, вызвав определенный метод библиотеки. 

Все библиотеки анализа данных Python делятся на четыре группы, в которых представлены библиотеки, позволяющие решать задачи разной уровни сложности: 

### Вычислительные библиотеки

Первая группа называется научными **вычислительными библиотеками**. Сюда относится библиотека **Pandas**. Она является одной из частоиспользуемых библиотек при работе с большими данными. Библиотека Pandas предполагает работу со структурами данных и предоставляет инструментарий эффективного манипулирования данными и их анализа.  

<img src="https://github.com/SerjiEvg/data-analysis/raw/main/images/pandas_logo.png" width="200"/>


Pandas отлично проявляет себя в сферах статистики, анализа данных, экономики, финансов. **Основная задача библиотеки - очистка и первичная оценка данных по общим показателям**. Инструментом библиотеки является двухмерная таблица, состоящая из меток столбцов и строк и называя **фреймом данных (DataFrame)**. Имеется поддержка объекта DataFrame, предназначенного для работы с индексированными массивами и позволяющий работать с различными форматами данных: начиная от CSV до TSV.

Библиотека уверенно справляется с обработкой больших объёмов данных и при необходимости использует возможности NumPy и SciPy. Впрочем именно взаимодополнение этих пакетов и привело к тому, что чаще всего их используют вместе. Pandas позволяет:

- Обменивать данные между структурами в памяти и файлами различных форматов;
- Переформатировать наборы данных (даже создавать сводные таблицы);
- Выполнять трёхэтапные операции типа «разделение, изменение, объединение»;
- Работать с временными рядами: формирование временных периодов и изменение интервалов.

**Библиотека NumPy** - это один из основных пакетов программного обеспечения научной библиотеки приложений для обработки больших многомерных массивов и матриц, большой набор расширенных математических функций и методов реализации которых, позволяет этим объектам выполнять операции. Лучшая и самая важная особенность Numpy — **интерфейс массива**.

<img src="https://github.com/SerjiEvg/data-analysis/raw/main/images/numpy_logo.png" width="200"/>

Его можно расширить до объектов для матриц и с небольшим изменением кода разработчики быстрее выполняют обработку. Этот интерфейс можно использовать для выражения изображений, звуковых волн и других двоичных необработанных потоков в виде массива действительных чисел в n-мерном формате. В общем, библиотека полезна для full-stack разработчиков, занимающихся задачами машинного обучения.

В NumPy (Numerical Python) встроено множество вычислительных механизмов. К таким особенностям можно отнести:

- Возможность использования и выполнения простых и достаточно сложных научных расчётов;
- Поддержка многомерных массивов;
- Множество встроенных методов для выполнения различных вычислений с большими и многомерными массивами данных;
- Возможность преобразования данных;
- Поддержка не только с числовых, но и других типов данных.

**Библиотека SciPy** - эта библиотека построена на базе NumPy. SciPy отлично подходит для выполнения статистических расчётов и поддерживает те же виды данных, что и библиотека NumPy. Можно сказать, что SciPy - это как "матрешка в матрешке": набор подпакетов, в которых реализованы различные вычислительные механизмы.

<img src="https://github.com/SerjiEvg/data-analysis/raw/main/images/scipy_logo.png" width="200"/>

 
SciPy предоставляет все **эффективные численные процедуры**, такие как оптимизация, численное интегрирование и многие другие, с использованием специальных подмодулей. Например, быстрое преобразование Фурье, обработку изображений, решение дифференциальных уравнений, механизмы линейной алгебры и так далее. Благодаря дополнительным пакетам SciPy может решать задачи, которые "не по зубам" NumPy.

SciPy использует массивы NumPy в качестве базовой структуры данных и поставляется с модулями для различных и частоиспользуемых задач в научном программировании. 

#### Библиотеки визуализации данных

**Библиотеки для визуализации данных** - лучший способ визуально показать содержательные результаты анализа. Эти библиотеки позволяют создавать графики, диаграммы и различные карты. 

**Пакет Matplotlib** - это низкоуровневая библиотека для создания 2D-графиков. С его помощью вы можете создавать различные графики: от гистограмм и диаграмм рассеяния до диаграмм Фейделя. 

<img src="https://github.com/SerjiEvg/data-analysis/raw/main/images/matplotlib_logo.png" width="200"/>

Библиотека дружелюбна по отношению к другим Python-пакетам, так что вы сможете легко использовать ее вместе с теми же NumPy и SciPy. Более того, Matplotlib поддерживает API, который позволяет встраивать графические объекты в различные приложения. Графические представления данных, создаваемые этим пакетом, поддаются гибкой настройке. Matplotlib поддерживает многие виды графиков и диаграмм, такие как:

- Графики (line plot);
- Диаграммы разброса (scatter plot);
- Столбчатые диаграммы (bar chart) и гистограммы (histogram);
- Круговые диаграммы (pie chart);
- Ствол-лист диаграммы (stem plot);
- Контурные графики (contour plot);
- Поля градиентов (quiver);
- Спектральные диаграммы (spectrogram) и т.д.

**Библиотека Seaborn** - это API высокого уровня, основанный на библиотеке matplotlib. Она содержит настройки по умолчанию, которые больше подходят для обработки графиков. Кроме того, существует множество библиотек визуализации, включая некоторые сложные типы, такие как временные ряды, диаграммы распределения и скрипичные диаграммы.

<img src="https://github.com/SerjiEvg/data-analysis/raw/main/images/seaborn_logo.png" width="200"/>

Помимо визуализации, эта библиотека неплохо справляется с анализом и исследованием данных. Методы библиотеки позволяют выполнять:

Визуализацию распределения метрических переменных, где используются такие графики, как distplot, jointplot, rugplot, kdeplot.
Визуализацию относительно распределения между парами переменных ( методы PairGrid, pairplot, FacetGrid).
Создание сложных графиков моделей линейной регрессии.

**Библиотека Plotly** - это популярная библиотека, которая позволяет легко создавать сложную графику. Пакет программного обеспечения подходит для интерактивных веб-приложений и может обеспечивать визуальные эффекты, такие как контурные диаграммы, троичные диаграммы и трехмерные диаграммы.

<img src="https://github.com/SerjiEvg/data-analysis/raw/main/images/plotly_logo.png" width="200"/>

Эта библиотека позволяет создавать интерактивные графики для последующего анализа взаимоотношений переменных. Пожалуй, ключевым отличием Plotly от Matplotlib является более глубокое, продвинутое построение трёхмерных графиков. Основные возможности:

Визуализация всех необходимых видов графиков: 
- линейные диаграммы, круговые диаграммы, пузырьковые диаграммы, точечные диаграммы, древовидные диаграммы;
- Трехмерная визуализация;
- Подготовка интерактивных отчетов и экспорт в формате JSON.

**Пакет Bokeh** - использует виджеты JavaScript для создания интерактивных и масштабируемых визуализаций в браузере. Библиотека предоставляет множество коллекций диаграмм, возможностей для стилизации, добавления виджетов, определения обратных вызовов и других форм интерактивных возможностей, а также множество других полезных функций. 

<img src="https://github.com/SerjiEvg/data-analysis/raw/main/images/boken_logo.png" width="200"/>
 

Bokeh легко интегрируется с любым фреймворком на Python, поддерживает много видов графиков, но самое главное - она позволяет выводить диаграммы и визуальные элементы JavaScript в веб-приложениях.

Да, это можно сделать и в D3.js, но для этого потребуются дополнительные инструменты. В случае Bokeh вы можете просто писать свой код на Python, а библиотека автоматически сгенерирует JavaScript для сайта или приложения. Это открывает огромные интерактивные возможности:

- Поддерживает визуализацию наборов данных, которые обычно используются в статистике и науке;
- Поддерживает различные форматы выходных данных;
- Существуют версии Bokeh для разных языков программирования;

Пакет хорошо интегрируется с такими Python-фреймворками, как Django и Flask.

#### Алгоритмические библиотеки

Алгоритмические библиотеки решают задачи машинного обучения различной уровни сложности. Мы же рассмотрим два пакета. 

**Библиотека Scikit-learn** содержит инструменты для статистического моделирования и решения задач машинного обучения, включая регрессию, классификацию, кластеризацию и т. д.  Это библиотека Python, связанная с NumPy и SciPy. Считается одной из лучших библиотек для работы с комплексными данными. Она построена на библиотеках NumPy, SciPy и Matplotlib. 

<img src="https://github.com/SerjiEvg/data-analysis/raw/main/images/scikit_learn_logo.png" width="200"/>

Scikit-learn - один из самых популярных пакетов для машинного обучения на Python. Встроенные в него модули дают возможность работать с множеством популярных алгоритмов машинного обучения (Machine Learning): алгоритм спектральной кластеризации, кросс-валидация, случайный лес (Random forest), k-средние (K-Means) и так далее. Scikit-learn можно использовать для создания моделей обучения с учителем (Supervised learning) и без.

При этом Scikit-learn абсолютно бесплатный даже для коммерческого использования. Открытый исходный код буквально лежит на GitHub и ждет вас. А чтобы не осталось никаких сомнений - перечислим, какие задачи поддерживает библиотека:

- Предварительная обработка данных;
- Уменьшение размерности;
- Выбор модели;
- Регрессии и классификации;
- Задачи кластеризации.

**Библиотека Statsmodels** - это модуль Python, который предоставляет множество возможностей для статистического анализа данных, таких как оценка статистической модели и выполнение статистических тестов. С его помощью вы можете реализовать множество методов машинного обучения и изучить возможности рисования. 

<img src="https://github.com/SerjiEvg/data-analysis/raw/main/images/statsmodel_logo.png" width="200"/>

Пакет выпущен под лицензией Modified BSD с открытым исходным кодом. С его интеграцией не придется ломать голову: для этого можно использовать упомянутый выше Pandas. StatsModels позволяет работать с

- Линейной регрессией;
- Обобщенными линейными моделями и оценочными уравнениями;
- Регрессией с дискретной зависимой переменной;
- Анализом временных рядов и временных моделей методами пространства состояний statespace;
- Таблицами непредвиденных обстоятельств;
- Множественными расчетами с цепными уравнениями.

#### Библиотеки машинного обучения

**Библиотека Keras** - это высокоуровневая библиотека для обработки нейронных сетей, которая работает на TensorFlow и Theano. Теперь, благодаря выходу новой версии, CNTK и MxNet также могут использоваться в качестве бэкэндов. Это упрощает многие конкретные задачи и значительно уменьшает объем монотонного кода. Тем не менее, он может не подходить для некоторых сложных задач.

<img src="https://github.com/SerjiEvg/data-analysis/raw/main/images/keras.png" width="200"/>

Эта библиотека содержит многочисленные реализации широко применяемых строительных блоков нейронных сетей: слои, целевые и передаточные функции, оптимизаторы, и множество инструментов для упрощения работы с изображениями и текстом.

Вот некоторые возможности данной библиотеки:

- Поддерживает широкий набор нейросетевых моделей;
- Содержит встроенные средства для работы с изображениями;
- Поддерживает популярные алгоритмы глубокого обучения;
- Отличается высоким уровнем расширяемости, что позволяет, при необходимости, оснащать его новым функционалом.

**Библиотека TensorFlow** - популярная среда глубокого и машинного обучения, разработанная Google Brain. Это дает возможность использовать искусственные нейронные сети с несколькими наборами данных. 
Среди наиболее популярных приложений TensorFlow - распознавание целей и распознавание речи. На обычном TensorFlow также построены библиотеки tflearn, tf-slim, skflow и т.д.

<img src="https://github.com/SerjiEvg/data-analysis/raw/main/images/tensor_flow_logo.png" width="200"/>

TensorFlow предоставляет надежные и масштабируемые решения для вычислений на множестве машин и для вычислений с огромными наборами данных. Следовательно, это предпочтительная структура для машинного обучения. Библиотека расширяема и поддерживает множество платформ. Она обеспечивает поддержку графического процессора для более быстрых вычислений, повышения производительности и лучшей визуализации. 
TensorFlow предоставляет алгоритмы для классификации, моделей оценки, дифференциации и т.д., а также поддерживает API для обучения нейронных сетей и распознавания речи с использованием NLP (обработка естественного языка).
 

**Библиотека PyTorch** - это большая структура, которая позволяет ускорению графического процессора выполнять тензорные вычисления (преобразование элементов одного линейного пространства в элементы другого), создавать динамические расчетные графики и автоматически вычислять градиенты. Помимо этого, PyTorch предоставляет богатый API для решения приложений, связанных с нейронными сетями. 
Библиотека основана на Torch и является библиотекой глубокого обучения с открытым исходным кодом, реализованной на C.
